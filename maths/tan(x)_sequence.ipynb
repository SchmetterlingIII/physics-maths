This code will try to find, iteratively, what the maximium of this sequence is such that:

$$x_{n+1} = \tan(x_n)$$

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

x = [1]
iteration_steps = 10000
for i in range(iteration_steps):
    y = np.tan(np.tan(x[-1]))
    x.append(y)

plt.scatter(np.linspace(0, iteration_steps+1, iteration_steps+1), x, s=0.3)
plt.grid(alpha=0.2)
plt.ylim(-50,50)

df = pd.DataFrame({"f(x)":x})
print(df)
```
